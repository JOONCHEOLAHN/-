## 토큰화(Tokenization)

```python
df_fin = df_fin['text'].astype(str)
from nltk.tokenize import word_tokenize  
word_tokenize(df_fin)
```

```python
import re
# def cleansing(txt):
sen_list = []
for sen in df_fin['text']:
    pattern = '(\[a-zA-Z0-9\_.+-\]+@\[a-zA-Z0-9-]\+.\[a-zA-Z0-9-.\]+)'  # 이메일 주소 제거
    txt = re.sub(pattern=pattern,repl=' ', string=txt)
    
    pattern = '(http|ftp|https)://(?:[-w.]|(?:\da-fA-F]{2}))+'          # url제거
    txt = re.sub(pattern=pattern,repl=' ', string=txt)
    
    pattern = '[\r|\n]'                                                 # \r, \n 제거
    txt = re.sub(pattern=pattern,repl=' ', string=txt)
    
    pattern = '[^\w|\s]'                                                # 특수기호 제거   
    txt = re.sub(pattern=pattern,repl=' ', string=txt)
    
    pattern = re.compile(r'\s+')                                        # 이중space 제거   
    txt = re.sub(pattern=pattern,repl=' ', string=txt)
    
#     return(txt)
```

```python
import re
from tensorflow.keras.preprocessing.text import text_to_word_sequence

sen_list = []
for sen in df_fin['text']:
    sen = re.sub(r'\([^)]*\)', '', sen.strip())  #(s) 제거
    sen = re.sub(r'[^)]*\)', '', sen.strip())    #s) 제거
    sen = re.sub(r'\([^)]*', '', sen.strip())    #(s 제거
    sen = re.sub(' +', ' ', sen.strip())       #공백,탭 제거
#     print(sen)
    #nous = kkma.nouns(sen)
    nous = text_to_word_sequence(sen)
    sen_list.append(nous)  
sen_list = sum(sen_list, [])  #2차원-->1차원으로 변환
pd.DataFrame(sen_list).value_counts().sort_values(ascending=False).head()    
```
